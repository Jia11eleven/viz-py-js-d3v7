```python
import pandas as pd
import json
from io import StringIO
```

```python
import os
path = "D:/Python_Scripts/DataViz_JS_Python/nobel_winners"
os.chdir(path)
os.getcwd()
```

```python
data=[]
with open('winners_cleaned.json', 'r') as file:
    for line in file:
        data.append(pd.read_json(StringIO(line), typ='series'))
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.date_of_birth = pd.to_datetime(df.date_of_birth, unit='ms', origin='unix',utc=True)
df['date_of_birth'] = df['date_of_birth'].dt.tz_localize(None)
df.date_of_death = pd.to_datetime(df.date_of_death)
df.head()
```

```python
for name,group in df.groupby('country'):
    group.to_json(name + '.json',
                  orient='records')
```
