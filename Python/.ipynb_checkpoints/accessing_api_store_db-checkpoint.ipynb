```python
#accessing MongoDB database
def get_mongo_database(db_name,host='localhost',
                       port=27017,username=None, password=None):
    '''get named database from MongoDB w/o authentication'''
    #make Mongo connection w/o authentication
    if username and password:
        #URI: uniform resource identifier)
        mongo_uri = 'mongodb://%s:%s@%s/%S'%(username, password,host, db_name)
        conn = MongoClient(mongo_uri)
    else:
        conn = MongoClient(host,port)
    return conn[db_name]
```

```python
db_nobel = get_mongo_database('nobel_prize')
col = db_nobel['country_data'] #country data collecion
```

```python
'''get all the RESTful country data, direct call will result in
brocken connection due to the large amount of data'''

response = REST_country_request(stream=True)
#insert the JSON objects straight to our collection
col.insert(response.json())
```

```python
#walk around solution, to split to smaller dataset
regions = ["Africa", "Asia", "Europe", "Americas", "Oceania"]
all_countries = []

for region in regions:
    try:
        response = requests.get(f"https://restcountries.com/v3.1/region/{region}", timeout=30)
        response.raise_for_status()
        all_countries.extend(response.json())
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data for region {region}: {e}")
```

```python
col.insert_many(all_countries)
```

```python
#currency is a dictionary
res = col.find({'currencies.USD':{'$exists':True}})
list(res)
```

```python
scope = [
    'https://www.googleapis.com/auth/spreadsheets',  # Full access to spreadsheets
    'https://www.googleapis.com/auth/drive'         # Access to Google Drive
]
credentials = ServiceAccountCredentials.from_json_keyfile_name(
    './data/gspread_credentials.json', scope)
gc = gspread.authorize(credentials)
ss = gc.open('Microbe-scope') 
#name of the spreadsheet, we can also use open_by_url / open_by_id
```
