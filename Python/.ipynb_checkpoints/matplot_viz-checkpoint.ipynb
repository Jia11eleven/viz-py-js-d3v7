```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = pd.period_range(datetime.now(),periods=200, freq='d')
x = x.to_timestamp().to_pydatetime()
y = np.random.randn(200,3).cumsum(0)
```

```python
plt.plot(x,y)
```

```python
import matplotlib as mpl
mpl.rcParams['lines.linewidth'] = 2
plt.plot(x, y[:, 0], label="Series 1", color='r')  # Red line
plt.plot(x, y[:, 1], label="Series 2", color='g')  # Green line
plt.plot(x, y[:, 2], label="Series 3", color='b')  # Blue line
```

```python
plt.rcParams['figure.figsize'] = (8,4)
plt.gcf().set_size_inches(8,4)
```

```python
plots = plt.plot(x,y)
plt.legend(plots,('foo','bar','baz'),
           loc='best',
           framealpha=0.5,
           prop={'size':'small','family':'monospace'})
plt.title('random trends')
plt.xlabel('Date')
plt.ylabel('Cum.sum')
plt.figtext(0.995, 0.01,'Acme designs 2015',
            ha='right', va='bottom')
plt.grid(True)
plt.tight_layout() #prevent plot elements from being obscured or truncated
plt.savefig('mpl_3lines_custom.svg')
plt.show()
```

Object-Oriented Matplotlib

```python
fig = plt.figure(figsize=(8,4), #in inches
                 dpi=200, #dots per inch
                 tight_layout=True, #fit to canvas
                 linewidth=1,
                 edgecolor='r')
#main axes
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.set_title('Main axes with inserted child axies')
ax.plot(x,y[:,0])
ax.set_xlabel('Date')
ax.set_ylabel('Cum.sum')
#inserted axes
ax = fig.add_axes([0.15, 0.15, 0.3, 0.3])
ax.plot(x,y[:,1], color='g')
ax.set_xticks([]);
```

```python
figure, ax = plt.subplots()
plots = ax.plot(x,y, label='')
figure.set_size_inches(8,4)
ax.legend(plots,('foo','bar','baz'),
          loc='best', framealpha=0.25,
          prop={'size':'small','family': 'monospace'})
ax.set_title('Random trends')
ax.set_xlabel('Date')
ax.set_ylabel('Cum.sum')
ax.grid(True)
figure.text(0.995, 0.011, u'\u00a9 Acme Designs 2015',
            ha='right',va='bottom')
figure.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1,
                         sharex=True, sharey=True,
                         figsize=(8,8))
labelled_data = zip(y.transpose(),
                    ('foo','bar','baz'),
                    ('b','g','r'))
fig.suptitle('Three Random Trends', fontsize=16)
for i, ld in enumerate(labelled_data):
    ax = axes[i]
    ax.plot(x, ld[0], label=ld[1], color=ld[2])
    ax.set_ylabel('Cum.sum')
    ax.legend(loc='upper left',framealpha=0.5, 
                prop={'size':'small'})
    axes[-1].set_xlabel('Date')
```

```python
labels = ['Physics','Chemistry','Literature','Peace']
foo_data = [3,6,10,4]
bar_width = 0.2
xlocations = np.array(range(len(foo_data)))+bar_width
plt.bar(xlocations,foo_data, width=bar_width)
plt.yticks(range(0,12))
plt.xticks(xlocations+bar_width/8,labels)
plt.xlim(0,xlocations[-1]+bar_width*2)
plt.gca().get_xaxis().tick_bottom()
plt.gca().get_yaxis().tick_left()
plt.gcf().set_size_inches((8,4))
```

```python
labels = ['Physics','Chemistry','Literature','Peace']
foo_data = [3,6,10,4]
bar_data=[8,3,6,1]

fig, ax = plt.subplots(figsize=(8,4))
bar_width = 0.4
xlocs = np.arange(len(foo_data))
ax.bar(xlocs-bar_width, foo_data, bar_width, 
       color = '#fde0bc', label = 'Fooland')
ax.bar(xlocs, bar_data, bar_width,
       color='peru', label='Barland')
#ticks, labels, frids and title
ax.set_yticks(range(12))
ax.set_xticks(ticks=range(len(foo_data)))
ax.set_xticklabels(labels)
ax.yaxis.grid(True)
ax.legend(loc='best')
ax.set_ylabel('Number of prizes')
fig.suptitle('Prizes by country')
fig.tight_layout(pad=2)
fig.savefig('mpl_barchart_multi.png', dpi=200)
```

```python
labels = ['Physics','Chemistry','Literature','Peace']
foo_data = [3,6,10,4]
bar_data=[8,3,6,1]

fig, ax = plt.subplots(figsize=(8,4))
bar_width = 0.4
ylocs = np.arange(len(foo_data))
ax.barh(ylocs-bar_width, foo_data, bar_width, 
       color = '#fde0bc', label = 'Fooland')
ax.barh(ylocs, bar_data, bar_width,
       color='peru', label='Barland')
#ticks, labels, frids and title
ax.set_xticks(range(12))
ax.set_yticks(ticks=range(len(foo_data)))
ax.set_yticklabels(labels)
ax.xaxis.grid(True)
ax.legend(loc='best')
ax.set_xlabel('Number of prizes')
fig.suptitle('Prizes by country')
fig.tight_layout(pad=2)
fig.savefig('mpl_barchart_multi_2.png', dpi=200)
```

```python
labels = ['Physics','Chemistry','Literature','Peace']
foo_data = [3,6,10,4]
bar_data=[8,3,6,1]

fig, ax = plt.subplots(figsize=(8,4))
bar_width = 0.8
xlocs = np.arange(len(foo_data))
ax.bar(xlocs+0.2, foo_data, bar_width, 
       color = '#fde0bc', label = 'Fooland')
ax.bar(xlocs+0.2, bar_data, bar_width,
       color='peru', label='Barland',
       bottom=foo_data)
#ticks, labels, frids and title
ax.set_yticks(range(18))
ax.set_xticks(ticks=np.array(range(len(foo_data))) + bar_width/4)
ax.set_xticklabels(labels)
ax.set_xlim(-(1-bar_width),xlocs[-1]+1)
ax.xaxis.grid(True)
ax.legend(loc='best')
ax.set_xlabel('Number of prizes')
fig.suptitle('Prizes by country')
fig.tight_layout(pad=2)
fig.savefig('mpl_barchart_multi_3.png', dpi=200)
```

```python
num_points = 100
gradient = 0.5
x = np.array(range(num_points))
y = np.random.randn(num_points)*10+x*gradient
fig,ax = plt.subplots(figsize=(8,4))
ax.scatter(x,y)
fig.suptitle('a simple scatter plot')
```

```python
num_points = 100
gradient = 0.5
x = np.array(range(num_points))
y = np.random.randn(num_points)*10+x*gradient
fig,ax = plt.subplots(figsize=(8,4))
colors = np.random.rand(num_points)
size = (2+np.random.rand(num_points)*8)**2
ax.scatter(x,y, s=size, c=colors, alpha=0.5)
fig.suptitle('a simple scatter plotcwith color and size specified')
```

```python
num_points = 100
gradient = 0.5
x = np.array(range(num_points))
y = np.random.randn(num_points)*10+x*gradient
fig,ax = plt.subplots(figsize=(8,4))
ax.scatter(x,y, s=size, c=colors, alpha=0.5)
m, c = np.polyfit(x,y,1) #1 stands for degree
ax.plot(x, m*x +c)
fig.suptitle('a simple scatter plot with regression line')
```

```python
import seaborn as sns #relies on matplotlib
import matplotlib as mpl
```

```python
data = pd.DataFrame({'dummy x':x,'dummy y':y})
```

```python
data.head(10)
```

```python
sns.lmplot(x='dummy x', y='dummy y',
            data=data, height=4, aspect=2)
plt.tight_layout()
plt.savefig('mpl_scatter_seaborn.png')
```

```python
sns.lmplot(x='dummy x', y='dummy y',
           data=data, height=4, aspect=2,
           scatter_kws={"color":"slategray"},
           line_kws={"linewidth":2,
                     "linestyle":'--',
                     "color":"seagreen"},
           markers='D',
           ci=68) #confidence interval
plt.tight_layout()
plt.savefig('mpl_scatter_seaborn_1.png')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(tips, col="smoker")
g.map(plt.scatter,"total_bill","tip")
```

```python
pal = dict(Female='red', Male='blue')
g = sns.FacetGrid(tips,col='smoker',hue='sex',
                  hue_kws={'marker':["D","s"]},
                  palette=pal,height=4,aspect=1)
g.map(plt.scatter,"total_bill","tip",alpha=0.4)
g.add_legend()
```

```python
pal = dict(Female='red', Male='blue')
g = sns.FacetGrid(tips,col='smoker',row='time',
                  hue='sex',hue_kws={'marker':["D","s"]},
                  palette=pal,height=4,aspect=1)
g.map(sns.regplot,"total_bill","tip",
      scatter_kws={'alpha':0.4})
g.add_legend()
```

```python
pal = dict(Female='red',Male='blue')
sns.lmplot(x="total_bill", y="tip",hue="sex",
           markers=["D","s"],
           col="smoker", row="time", data=tips, 
           palette=pal, height=4, aspect=1)
```
